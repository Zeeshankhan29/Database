```python
class Mongo:
    def __init__(self,username,password):
        self.username = username
        self.password = password
        self.Myurl = f"mongodb+srv://{username}:{password}@cluster0.gc1mb.mongodb.net/?retryWrites=true&w=majority"
            
    def Myclient(self):
        client = pymongo.MongoClient(self.Myurl)
        return client
    
    def database(self,db_name):
        client = self.Myclient()
        db_name = client[db_name]
        return db_name
    
    def collection(self,db_name,col_name):
        db_name = self.database(db_name)
        col_created = db_name[col_name]
        return col_created
    
    def find_all(self,db_name,col_name):
        l=[]
        col_created = self.collection(db_name,col_name)
        for i in col_created.find():
            l.append(i)
        return l
        
    def findone(self,key,value,db_name,col_name):
        l=[]
        col_created = self.collection(db_name,col_name)
        for i in col_created.find_one({key:value}):
            l.append(i)
        return l
            
            
```

```python
class Mongodb:

    def __init__(self,username,password):
        self._a=username
        self.__b=password
        self.url=pymongo.MongoClient(f"mongodb+srv://{self._a}:{self.__b}@cluster0.gc1mb.mongodb.net/?retryWrites=true&w=majority")
        
    def client(self):
        return self.url
    
    def db_con(self,db_name):
        return self.client()[db_name]
    
    def col_con(self,db_name,col_name):
        return self.db_con(db_name)[col_name]
    
    def insert(self,db_name,col_name,key,value):
        var=self.col_con(db_name,col_name)
        var1=var.insert_one({key:value})
        return var1
    
    def find(self,db_name,col_name):
        l=[]
        var=self.col_con(db_name,col_name)
        for i in var.find():
            l.append(i)
        return l
```
