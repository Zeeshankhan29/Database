```python
class xyz:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
    
    def test1(self):
        return 'this is the test1 method of class xyz'
    
    def test2(self):
        return 'this is the test2 method of class xyz'
    
class xyz1(xyz):
    def __init__(self,p,q,r,*args):
        xyz.__init__(self,*args)
        self.p=p
        self.q=q
        self.r=r
    
    def test3 (self):
        return 'this is the test3 method of class xyz1'
    
    def test4(self):
        return 'this is the test4 method of class xyz1'
    
class xyz2(xyz1):
    def __init__(self,u,v,w,*args):
        xyz1.__init__(self,*args)
        self.u=u
        self.v=v
        self.w=w
   

    def test5(self):
        return 'this is the test5 method of class xyz1'

    def test6(self):
        return 'this is the test6 method of class xyz1'    
    
        
```

```python
a=[[[[1,23,4,56,7]]]]
b=[[[[23,250,40,560,70]]]]
```

```python
l=[]
q=[]
for i in a[0][0][0]:
    l.append(i)
for j in b[0][0][0]:
    q.append(j)
for m in l:
    for n in q:
        print(m-n,end=',')
```
