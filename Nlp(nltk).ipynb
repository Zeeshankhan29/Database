```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from nltk import word_tokenize
```

```python
#tokenization -- convert paragraph to sentences and focus on words
nltk.download('punkt')
sentences = nltk.sent_tokenize(corpus)
```

```python
#Applying lemmatization
lem = WordNetLemmatizer()
```

```python
lem.lemmatize('goes')
```

```python
corpus=[]
import re
for i in range(len(sentences)):
    review = re.sub('[^a-zA-Z]', ' ',sentences[i])
    review = review.lower()
    review = review.split()
    review = [lem.lemmatize(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus 
```

```python
#Stemming

ob = PorterStemmer()
for i in corpus:
    words = word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(ob.stem(word))
```

```python
#lemmatization

ob = WordNetLemmatizer()
for i in corpus:
    words = word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(ob.lemmatize(word))
```

# Flow How Nlp works

```python
#Tokenization (converting sentences to words)
a = 'i am not well today so going to office is cancelled'
from nltk import word_tokenize
words = word_tokenize(a)
words
```

```python
#Applying stopwords to remove unmeaning words
from nltk.corpus import stopwords
j=[]
for word in words:
    if word not in set(stopwords.words('english')):
        j.append(word)
print(j)
```

```python
#Applying stemming (gives the base word)
from nltk.stem import PorterStemmer
ob = PorterStemmer()
for i in range(len(j)):
    print(ob.stem(j[i]))
```

```python
#Applying lemitization
from nltk.stem import WordNetLemmatizer
ob = WordNetLemmatizer()
for i in range(len(j)):
    print(ob.lemmatize(j[i]))
```
